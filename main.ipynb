```python
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re
```

## 1. Reading Data:

```python
with open('captions.txt', 'r') as f:
    next(f)                             # to skip the header
    contents = f.read().split('\n')

f.close()
contents
```

## 2. Cleaning the Text-Data:

```python
#For each Image we have 5 captions, we make dictionary for each image, at the same time we clean the data:

stop_words = set(stopwords.words('english'))

def tokenizer(s):
        s.lower
        words = nltk.tokenize.word_tokenize(s)
        words = [t for t in words if len(t) >= 2]  # remove short words
        words = [t for t in words if t not in stop_words]  # remove stopwords
        words = [t for t in words if not any(c.isdigit() for c in t)]
        sent= ' '.join(map(str,words))       #full cleaned sentence
        
        return words , sent



cap_dict = {}              #key: number of Image , Value: descriptions
all_tokens = []
unique_tokens = []

#### make this dictionary to count the frequency of each token##############
word_index_map={}
current_index = 0

############################################################################
for item in contents:
    item = item.split(',')
    tokens, description = tokenizer(item[-1])        
    
##### make a dictionary for image and corresponding descriptions ##########
    if item[0] in cap_dict:
        cap_dict[item[0]].append(description)
    else:
        cap_dict[item[0]] = [description]

##### make a list of vocabulary ############################################
    for token in tokens:
        all_tokens.append(token)

#### Indexing each tokens ##################################################     
        if token not in word_index_map:
            unique_tokens.append(token)
            word_index_map[token] = current_index        
            current_index += 1

    
cap_dict
```

```python
########## Here We count the tokens and remove outliers from vocabulary

counter_array = np.zeros(len(word_index_map), int)
less_frequent_words =[]

for item in all_tokens:
    i = word_index_map[item]
    counter_array[i] += 1

result = np.where(counter_array <= 10)

for x in result[0]:
    word= list(word_index_map.keys())[list(word_index_map.values()).index(x)]
    less_frequent_words.append(word)

print("From total number of {} tokens, {} are repeated lower than 10 times".format(len(all_tokens), len(less_frequent_words)))
```

```python
for item in less_frequent_words:
    all_tokens.remove(item)
    unique_tokens.remove(item)
    

print(len(all_tokens))
len(unique_tokens)
```
